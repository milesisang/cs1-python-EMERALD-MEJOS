```python
def display_assessment_analysis(assessment_name, scores_list):
  """
  Calculates and displays the sum and average of a list of scores for a given assessment.
  """
  print(f"\n--- {assessment_name} Analysis Results ---")
  if scores_list:
    total_sum = sum(scores_list)
    average = total_sum / len(scores_list)
    print(f"Total sum of {assessment_name.lower()} scores: {total_sum}")
    print(f"Average of {assessment_name.lower()} scores: {average:.2f}")
  else:
    print(f"No valid {assessment_name.lower()} scores were entered.")
  print("--------------------------------")
```

```python
def run_multiple_assessments():
  """
  Collects scores for multiple assessment types (assessments, quizzes, exams)
  and displays a single combined weighted final grade.
  """
  print("Starting multiple assessment score collection...")

  assessment_scores = get_numbers_for_assessment("assessment")
  quiz_scores = get_numbers_for_assessment("quiz")
  exam_scores = get_numbers_for_assessment("exam")

  final_weighted_grade_value = 0
  total_weights_applied = 0

  if assessment_scores:
    avg_assessment = sum(assessment_scores) / len(assessment_scores)
    final_weighted_grade_value += avg_assessment * 0.30
    total_weights_applied += 0.30

  if quiz_scores:
    avg_quiz = sum(quiz_scores) / len(quiz_scores)
    final_weighted_grade_value += avg_quiz * 0.30
    total_weights_applied += 0.30

  if exam_scores:
    avg_exam = sum(exam_scores) / len(exam_scores)
    final_weighted_grade_value += avg_exam * 0.40
    total_weights_applied += 0.40

  # Normalize the grade if some categories were not entered, based on weights applied
  if total_weights_applied > 0:
    final_weighted_grade_value /= total_weights_applied
  else:
    final_weighted_grade_value = 0 # No scores entered at all

  # Display combined analysis. Since get_numbers_for_assessment returns floats (e.g., 0.5 for 1/2),
  # the final_weighted_grade_value will be between 0 and 1. Multiply by 100 for percentage display.
  # display_assessment_analysis expects a list, so wrap the single grade in a list.
  display_assessment_analysis("Final grade", [final_weighted_grade_value * 100])

  print("\nAll assessment analyses complete!")
```

```python
def get_numbers_for_assessment(assessment_type_prompt):
  """
  Prompts the user to enter a specified number of scores for an assessment
  and returns a list of valid numerical scores (can be integers or fractions).
  """
  while True:
    try:
      count_str = input(f"How many {assessment_type_prompt} scores do you want to enter?: ")
      num_count = int(count_str)
      if num_count <= 0:
        print("Please enter a positive whole number for the count.")
        continue
      break
    except ValueError:
      print("Invalid input. Please enter a whole number for the count.")

  scores_list = []

  print(f"\nPlease enter {num_count} numbers for {assessment_type_prompt} (e.g., 38/50):")

  for i in range(num_count):
    while True:
      num_input = input(f"Enter number {i + 1}: ")
      try:
        if '/' in num_input:
          numerator_str, denominator_str = num_input.split('/')
          num = float(numerator_str) / float(denominator_str)
        else:
          num = float(num_input)

        if num < 0:
          print("Score cannot be negative. Please enter a positive number.")
          continue

        scores_list.append(num)
        break
      except ValueError:
        print("Invalid input. Please enterfraction (e.g., (38/50).")
      except ZeroDivisionError:
        print("Denominator cannot be zero. Please enter a valid fraction.")
  return scores_list
```

```python
run_multiple_assessments()
```
